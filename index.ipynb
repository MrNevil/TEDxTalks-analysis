## 2. Which Talks provoke the most online discussion?

```python
ted.sort_values('comments').tail()
```

If we analyze by just using the comments, there are few limitations such as:
1. The subcomments of a comment aren't accessible in our dataset.
2. How long it is online (since it is relevent)? 
Some talks have been online over a decade, others have been online for a day. So we can normalise this by using comments/views (which would give 0.__ values). 

To avoid smaller, therefore to make it more interpretable, we invert the ratio to views/comments.

```python
ted['comments_per_view'] = ted.comments / ted.views
```

```python
ted.sort_values('comments_per_view').tail()
```

```python
ted['views_per_comment'] = ted.views / ted.comments
```

```python
ted.sort_values('views_per_comment').head()
```

Lessons:

1. Consider the limitations and biases of your data when analyzing it
2. Make your results understandable (The reason why we used views/comments)
